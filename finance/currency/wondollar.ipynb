```python
import os
from fredapi import Fred
import matplotlib.pyplot as plt
import pandas as pd

# FRED API 키 가져오기
FRED_API_KEY = os.getenv('FRED_API')
fred = Fred(api_key=FRED_API_KEY)

# 원-달러 환율 데이터 가져오기
wondollar_data = fred.get_series('DEXKOUS')

# 날짜 필터링: 3년 전부터 오늘까지
end_date = pd.Timestamp.today()
start_date = end_date - pd.DateOffset(years=3)
filtered_data = wondollar_data[start_date:end_date]

# 평균값 계산
mean_value = filtered_data.mean()

# 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(filtered_data, label='Won-Dollar Exchange Rate', color='blue')
plt.axhline(y=mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
plt.title('Won-Dollar Exchange Rate (Last 3 Years)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid()
plt.show()
```
